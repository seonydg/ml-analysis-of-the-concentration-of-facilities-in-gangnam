```python
tmp = []
for i in tqdm(range(57)):
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    for j in range(5):
        brand = 'GS25'
        store = soup.select('#storeInfoList > tr')[j].text.split()[0][:-2]
        gu = '강남구'
        dong = soup.select('#storeInfoList > tr')[j].text.split('동')[0][-2:]

        try:
            address_1 = soup.select('#storeInfoList > tr')[j].text.split('(')[1].split(')')[0]
            address_2 = ' '.join(soup.select('#storeInfoList > tr')[j].text.split('(')[0].split()[1:])
        except:
            address_1 = ' '.join(soup.select('#storeInfoList > tr')[3].text.split('CAFE25')[0].split()[1:])
            address_2 = ' '.join(soup.select('#storeInfoList > tr')[3].text.split('CAFE25')[0].split()[1:])
        lat, lng = str(soup.select('#storeInfoList > tr')[j]).split('(\'map\', ')[1].split(');\"')[0].split(',')

        print(brand, store, gu, dong, address_1, address_2, lat, lng)

        tmp.append(
            {
                'brand':brand,
                'store':store,
                'gu':gu,
                'dong':dong+'동',
                'address_1':address_1,
                'address_2':address_2,
                'lat':lat,
                'lng':lng
            }
        )

    time.sleep(0.5)

    driver.find_element(By.CSS_SELECTOR, '#pagingTagBox > a.next').click()
    
```
